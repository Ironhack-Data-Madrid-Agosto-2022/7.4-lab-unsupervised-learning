```python
# Your import here:

from sklearn.preprocessing import StandardScaler

# Your code here:
scaler = StandardScaler()
```

```python
customers_scale = customers.copy() #copia por si acaso
```

```python
customers_scale[customers_scale.columns] = scaler.fit_transform(customers_scale[customers_scale.columns])
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

___

 Ahora agrupemos los datos con K-Means primero. Inicie el modelo K-Means, luego ajuste sus datos escalados. En los datos devueltos del método .fit, hay un atributo llamado etiquetas_ que es el número de clúster asignado a cada registro de datos. Lo que puede hacer es asignar estas etiquetas a los clientes en una nueva columna llamada clientes [ 'etiquetas' ]. Luego verá los resultados del clúster de los datos originales.

```python
from sklearn.cluster import KMeans

# Your code here:
kmeans=KMeans(n_clusters=5)
```

Count the values in `labels`.

```python
# Your code here:
clusters=kmeans.fit(customers_scale[customers_scale.columns]);
```

```python
preds=clusters.fit_predict(customers_scale[customers_scale.columns]);
```

```python
preds
```

```python
customers['labels'] = list(preds)
```

```python
customers.columns
```

```python
customers.groupby("labels").mean()
```

```python
customers.groupby("labels").median()
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

____

 Ahora agrupemos los datos usando DBSCAN. Use `DBSCAN ( eps = 0.5 )` para iniciar el modelo, luego ajuste sus datos escalados. En los datos devueltos del método `.fit`, asigne las `etiquetas_` nuevamente a `clientes [ 'labels_DBSCAN' ]`. Ahora sus datos originales tienen dos etiquetas, una de K-Means y la otra de DBSCAN.

```python
# Your code here
from sklearn.cluster import DBSCAN 
```

```python
dbscan=DBSCAN(eps=0.5)
```

Count the values in `labels_DBSCAN`.

```python
dbscan=DBSCAN(eps=0.5)
clusters=dbscan.fit(customers[customers.columns])
```

```python
preds=clusters.fit_predict(customers[customers.columns])
```

```python
labels = dbscan.labels_
customers["labels_DBSCAN"] = labels
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)).
